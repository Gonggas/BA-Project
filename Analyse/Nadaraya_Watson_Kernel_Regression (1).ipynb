## create center points and calculate distance to center

```python
# Calculate the center longitude and latitude for each grid cell
data['lon_center'] = (lon_bins[data['lon_bin']] + lon_bins[data['lon_bin'] + 1]) / 2
data['lat_center'] = (lat_bins[data['lat_bin']] + lat_bins[data['lat_bin'] + 1]) / 2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth's radius in kilometers
    lat1_rad, lon1_rad = np.radians(lat1), np.radians(lon1)
    lat2_rad, lon2_rad = np.radians(lat2), np.radians(lon2)
    dlat, dlon = lat2_rad - lat1_rad, lon2_rad - lon1_rad
    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c

# Calculate the distance from each data point to the center of its grid cell
# Make sure to pass the latitude first and then the longitude to the function
data['distance_to_center'] = haversine(data['Latitude'], data['Longitude'], data['lat_center'], data['lon_center'])

# data
```

```python
from statsmodels.nonparametric.kernel_regression import KernelReg

# Assuming 'residuals' is a column in your dataset representing the OLS residuals
# Aggregate data by grid cells, computing the mean 'residual' for each cell
grid_data = data.groupby(['lon_bin', 'lat_bin'])['residuals'].mean().reset_index()

# Calculate the center longitude and latitude for each grid cell that has data
grid_data['lon_center'] = (lon_bins[grid_data['lon_bin']] + lon_bins[grid_data['lon_bin'] + 1]) / 2
grid_data['lat_center'] = (lat_bins[grid_data['lat_bin']] + lat_bins[grid_data['lat_bin'] + 1]) / 2

# Perform Nadaraya-Watson kernel regression using the grid cell centers as independent variables
# and the mean residuals as the dependent variable
model_kw_resid = KernelReg(endog=grid_data['residuals'], exog=[grid_data['lon_center'], grid_data['lat_center']], var_type='cc', bw='cv_ls')
estimates_resid, marginal_effects_resid = model_kw_resid.fit()

# The 'estimates_resid' are the predicted residuals from the kernel regression
```

```python
import matplotlib.pyplot as plt

# Find the maximum absolute residual to set the limits of the colorbar symmetrically around zero
max_abs_resid = np.max(np.abs(estimates_resid))

# Plot the estimated residuals on the grid
plt.figure(figsize=(10, 10))
sc = plt.scatter(grid_data['lon_center'], grid_data['lat_center'], c=estimates_resid,
                 cmap='coolwarm', vmin=-max_abs_resid, vmax=max_abs_resid, label='Estimated Residuals')

# Add a color bar to show the residual scale, centered around zero
cbar = plt.colorbar(sc, label='Residuals')
cbar.set_label('Residuals', rotation=270)

# Add labels and title
plt.title('Nadaraya-Watson Kernel Regression Residual Estimates')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()

# Optionally, add grid lines
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Load the dataset
data = pd.read_csv('Clean_Data_17-04.csv')

# Make copies of the slices to avoid SettingWithCopyWarning
pre_lockdown_data = data[data['After corona lockdown'] == 0].copy()
post_lockdown_data = data[data['After corona lockdown'] == 1].copy()

# Decide on the number of cells in each dimension
num_cells_lon = 120
num_cells_lat = 120

# Create bins for longitude and latitude using the whole dataset to get the full range
min_lon, max_lon = data['Longitude'].min(), data['Longitude'].max()
min_lat, max_lat = data['Latitude'].min(), data['Latitude'].max()
lon_bins = np.linspace(min_lon, max_lon, num_cells_lon + 1)
lat_bins = np.linspace(min_lat, max_lat, num_cells_lat + 1)

# Assign each data point to a grid cell for both periods
for df in [pre_lockdown_data, post_lockdown_data]:
    df['lon_bin'] = np.digitize(df['Longitude'], lon_bins) - 1
    df['lat_bin'] = np.digitize(df['Latitude'], lat_bins) - 1
    df['lon_bin'] = df['lon_bin'].clip(0, num_cells_lon - 1)
    df['lat_bin'] = df['lat_bin'].clip(0, num_cells_lat - 1)
```

```python
# Prepare the data for OLS regression, excluding specified columns
predictors = [col for col in data.columns if col not in ['Date', 'Price', 'Status', 'URL', 'Adresse', 
                                                         'Kort adresse', 'Kvdm', 'Latitude', 'Longitude', 
                                                         'Kommune', 'Type', 'Price_per_kvdm', 'Price_million', 
                                                         'Year', 'Natural_log_Price', 'Natural_log_Price_per_kvdm',
                                                         'Nearest_Metro_Distance', 'Nearest_Tog_Distance', 'After corona lockdown','Toiletter']]

X_pre = sm.add_constant(pre_lockdown_data[predictors])
X_post = sm.add_constant(post_lockdown_data[predictors])

# The dependent variable for both pre and post lockdown data
Y_pre = pre_lockdown_data['Natural_log_Price_per_kvdm']
Y_post = post_lockdown_data['Natural_log_Price_per_kvdm']

# Fit the OLS model for pre-lockdown
model_pre = sm.OLS(Y_pre, X_pre)
results_pre = model_pre.fit()
pre_lockdown_data['residuals'] = results_pre.resid

# Fit the OLS model for post-lockdown
model_post = sm.OLS(Y_post, X_post)
results_post = model_post.fit()
post_lockdown_data['residuals'] = results_post.resid

# Now you have the residuals stored in pre_lockdown_data and post_lockdown_data DataFrames.


print("Pre-lockdown OLS regression results:")
results_pre.summary()
```

```python
print("\nPost-lockdown OLS regression results:")
results_post.summary()
```

```python
# Aggregate data by grid cells, computing the mean 'residual' for each cell
pre_lockdown_grid = pre_lockdown_data.groupby(['lon_bin', 'lat_bin'])['residuals'].mean().reset_index()
post_lockdown_grid = post_lockdown_data.groupby(['lon_bin', 'lat_bin'])['residuals'].mean().reset_index()

# Calculate grid cell centers
pre_lockdown_grid['lon_center'] = (lon_bins[pre_lockdown_grid['lon_bin']] + lon_bins[pre_lockdown_grid['lon_bin'] + 1]) / 2
pre_lockdown_grid['lat_center'] = (lat_bins[pre_lockdown_grid['lat_bin']] + lat_bins[pre_lockdown_grid['lat_bin'] + 1]) / 2
post_lockdown_grid['lon_center'] = (lon_bins[post_lockdown_grid['lon_bin']] + lon_bins[post_lockdown_grid['lon_bin'] + 1]) / 2
post_lockdown_grid['lat_center'] = (lat_bins[post_lockdown_grid['lat_bin']] + lat_bins[post_lockdown_grid['lat_bin'] + 1]) / 2
```

```python
# Assuming 'lon_bins' and 'lat_bins' have been previously created as per your grid definition

# Perform Nadaraya-Watson kernel regression on the pre-lockdown grid data
model_kw_pre = KernelReg(endog=pre_lockdown_grid['residuals'], 
                         exog=[pre_lockdown_grid['lon_center'], pre_lockdown_grid['lat_center']], 
                         var_type='cc', 
                         bw='cv_ls')  # cv_ls is cross-validation for bandwidth selection
estimates_pre, _ = model_kw_pre.fit()

# Perform Nadaraya-Watson kernel regression on the post-lockdown grid data
model_kw_post = KernelReg(endog=post_lockdown_grid['residuals'], 
                          exog=[post_lockdown_grid['lon_center'], post_lockdown_grid['lat_center']], 
                          var_type='cc', 
                          bw='cv_ls')
estimates_post, _ = model_kw_post.fit()

# Now, you can compare the estimates_pre and estimates_post to see the changes in the residuals pattern
```

```python
# Assuming 'estimates_pre' and 'estimates_post' are the results from NW kernel regression
# Find the maximum absolute residual to set the limits of the colorbar symmetrically around zero
max_abs_resid = max(np.max(np.abs(estimates_pre)), np.max(np.abs(estimates_post)))

# Set up subplots
fig, axes = plt.subplots(1, 2, figsize=(20, 10), sharey=True, sharex=True)

# Plot for pre-lockdown residuals
sc1 = axes[0].scatter(pre_lockdown_grid['lon_center'], pre_lockdown_grid['lat_center'], c=estimates_pre,
                      cmap='coolwarm', vmin=-max_abs_resid, vmax=max_abs_resid)
axes[0].set_title('Pre-Lockdown Estimated Residuals')
axes[0].set_xlabel('Longitude')
axes[0].set_ylabel('Latitude')
axes[0].grid(True)

# Plot for post-lockdown residuals
sc2 = axes[1].scatter(post_lockdown_grid['lon_center'], post_lockdown_grid['lat_center'], c=estimates_post,
                      cmap='coolwarm', vmin=-max_abs_resid, vmax=max_abs_resid)
axes[1].set_title('Post-Lockdown Estimated Residuals')
axes[1].set_xlabel('Longitude')
axes[1].grid(True)

# Create a single colorbar in the middle
fig.subplots_adjust(right=0.85)
cbar_ax = fig.add_axes([0.87, 0.15, 0.03, 0.7])  # Adjust these values to better fit your layout if needed
fig.colorbar(sc1, cax=cbar_ax, label='Residuals')

plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust to prevent overlap if necessary
plt.show()
```

```python
# Assuming the NW residuals 'estimates_pre' and 'estimates_post' have been calculated as shown above

# Map the NW smoothed residuals back to the original pre-lockdown and post-lockdown datasets
# Here, 'lon_bin' and 'lat_bin' are used to merge the grid-level NW residuals with the property-level data
pre_lockdown_grid['nw_residuals'] = estimates_pre
post_lockdown_grid['nw_residuals'] = estimates_post

# Merge the NW residuals into the original datasets based on grid cells
pre_lockdown_data = pre_lockdown_data.merge(pre_lockdown_grid[['lon_bin', 'lat_bin', 'nw_residuals']], 
                                            on=['lon_bin', 'lat_bin'], how='left')
post_lockdown_data = post_lockdown_data.merge(post_lockdown_grid[['lon_bin', 'lat_bin', 'nw_residuals']], 
                                              on=['lon_bin', 'lat_bin'], how='left')

# Prepare the data for the hybrid OLS regression
# Including NW estimates as predictors along with the original predictors
X_pre_hybrid = sm.add_constant(pre_lockdown_data[predictors + ['nw_residuals']])
X_post_hybrid = sm.add_constant(post_lockdown_data[predictors + ['nw_residuals']])

# The dependent variable remains the same for both periods
Y_pre = pre_lockdown_data['Natural_log_Price_per_kvdm']
Y_post = post_lockdown_data['Natural_log_Price_per_kvdm']

# Fit the hybrid OLS model for pre-lockdown
hybrid_model_pre = sm.OLS(Y_pre, X_pre_hybrid)
hybrid_results_pre = hybrid_model_pre.fit()

# Fit the hybrid OLS model for post-lockdown
hybrid_model_post = sm.OLS(Y_post, X_post_hybrid)
hybrid_results_post = hybrid_model_post.fit()

# Output the summaries of the hybrid models
print("Pre-lockdown hybrid OLS-NW regression results:")
hybrid_results_pre.summary()
```
