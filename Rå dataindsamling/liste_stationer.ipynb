## Scraper alle Metrostationer og deres koordinater

```python
import requests
from bs4 import BeautifulSoup
import re

# URL of the webpage
url = "https://danskejernbaner.dk/vis.stations.oversigt.php"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all elements with class "text-underline-hover"
    elements = soup.find_all(class_="text-underline-hover")

    # List to store the extracted text
    extracted_text = []

    # Iterate through the elements and extract text containing "metrostation"
    for element in elements:
        text = element.get_text().strip()
        if "metrostation" in text.lower():
            extracted_text.append(text)

    # Remove parentheses using regular expression
    extracted_text = [re.sub(r'\([^()]*\)', '', text) for text in extracted_text]

#     # Print the extracted text
#     for text in extracted_text:
#         print(text)
# else:
#     print("Failed to retrieve data from the website.")
```

```python
import pandas as pd
import googlemaps

# Function to geocode addresses and retrieve coordinates
def geocode_address(address):
    try:
        # Geocode address
        geocode_result = gmaps.geocode(address)
        if geocode_result:
            # Extract latitude and longitude from geocode result
            location = geocode_result[0]['geometry']['location']
            return location['lat'], location['lng']
        else:
            return None, None
    except Exception as e:
        print(f"Error geocoding address '{address}': {e}")
        return None, None

# Initialize Google Maps API client
gmaps = googlemaps.Client(key='AIzaSyAkzxBl6dpAhqaPG3WQEDPsaK3pRsJoMjw')  # Replace 'YOUR_API_KEY' with your actual API key

# List to store geocoded results
geocoded_results = []

# Geocode addresses and retrieve coordinates for each metro station
for metro_station in extracted_text:
    lat, lng = geocode_address(metro_station)
    geocoded_results.append((metro_station, lat, lng))
```

```python
# Create a DataFrame from the geocoded results
df = pd.DataFrame(geocoded_results, columns=['Metro', 'Latitude', 'Longitude'])

# Manually insert values for Latitude and Longitude
df.at[20, 'Latitude'] = 55.630440
df.at[20, 'Longitude'] = 12.648962

# Save the DataFrame as a CSV file
df.to_csv('metro_stations_geocoded.csv', index=False)

# Print the DataFrame
# print(df)
```

## Gør det samme for S-togstationer

```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://da.wikipedia.org/wiki/S-togsstationer"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all tables on the page
    tables = soup.find_all("table", class_="wikitable")

    # List to store station names
    stationer = []

    # Loop through each table to find the one containing station names
    for table in tables:
        # Find all table rows
        rows = table.find_all("tr")

        # Check if the first row contains "Station" header
        header_row = rows[0]
        header_cells = header_row.find_all("th")
        if header_cells and "Station" in header_cells[0].get_text():
            # Extract text from the first column of each row
            for row in rows[1:]:
                cells = row.find_all("td")
                if cells:  # Check if cells exist
                    first_column_text = cells[0].get_text(strip=True)
                    stationer.append(first_column_text)

            break  # Exit the loop after finding the desired table

#     if stationer:
#         print("Stationer list:")
#         for station in stationer:
#             print(station)
#     else:
#         print("No station names found on the page.")
# else:
#     print("Failed to retrieve data from the website.")
```
